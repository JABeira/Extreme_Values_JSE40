# Data Preproccessing

Our data, while intuitive, is not currently ideal for Extreme Value study. I propose the following transformations

## Log Returns
Raw asset price relatives ($PR_t$) are not additive but rather multiplicative: $$1 + R_{total} = (1 + R_1) \times (1 + R_2) \times \dots \times (1 + R_k)$$
While this is technically 'fine', it's computationally expensive to model over long horizons.
Applying a Natural log transform converts this product into a sum: $$r_{total} = r_1 + r_2 + \dots + r_k$$

Moreover, Simple returns suffer from asymmetry when dealing with gains and losses.If a stock price drops from 100 to 50, the simple return is -50%. To recover back to 100, the stock must gain 100%. This magnitude mismatch distorts variance.Log returns naturally correct this asymmetry.Log Loss: $\ln(50/100) = -0.693$Log Gain: $\ln(100/50) = +0.693$
Equal magnitude movements in the underlying asset price result in equal magnitude log returns.

### An interesting note from Gemini
Even though log returns solve the boundary and additivity problems, they do not solve the issue of volatility clustering. Extreme market losses tend to group together, violating the Independent and Identically Distributed (IID) assumption required by pure EVT.

## Invert the returns
We remain focused on estimating extreme market losses, we need to invert the returns so that large negative returns (market crashes) become large positive values in our right-tail EVT distributions. We define the daily loss sequence as:
$L_t = -r_t$


```python
import pandas as pd
import numpy as np
from pathlib import Path

file_path = '../data/JSE_top40_2003_2018.csv' 
df = pd.read_csv(file_path, parse_dates=['date'], index_col='date')

output_dir = Path("../data/asset_analysis")
output_dir.mkdir(exist_ok=True)
```

```python
for ticker in df.columns:

    pr_series = df[ticker]

    # Calculate log returns and losses
    log_return = np.log(pr_series)
    loss = -log_return

    # Create analysis DataFrame
    analysis_df = pd.DataFrame({
        "Price_Relative": pr_series,
        "Log_Return": log_return,
        "Loss": loss
    }).dropna()

    # Save to CSV
    output_file = output_dir / f"{ticker}_analysis.csv"
    analysis_df.to_csv(output_file)

    print(f"Saved {output_file}")

print(f"\nFinished processing {len(df.columns)} assets.")
```
