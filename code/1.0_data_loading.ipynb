# Data Interpretation

Dataset is price relatives ($PR_t$).
A value greater than 1 (e.g., 1.013216 for ACL on 2003-01-03) means the stock price increased by approximately 1.32% compared to the previous trading day.
Similarly, a value less than 1 (e.g., 0.957608 for ACL on 2003-01-06) means the stock price decreased by approximately 4.24%.
