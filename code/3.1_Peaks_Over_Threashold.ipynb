# The Peaks-Over-Threshold
This approach solves the data-wasting problem of Block Maxima. Instead of arbitrarily taking the single worst day of the year, POT extracts all daily losses that exceed a defined  threshold, $u$.If a daily loss $x$ is greater than our threshold $u$, we define the "exceedance" as $y = x - u$.The Balkema-de Haan-Pickands theorem states that for a sufficiently high threshold $u$, the distribution of these exceedances $y$ converges to the Generalized Pareto Distribution (GPD).The cumulative distribution function for the GPD is:$$G(y) = 1 - \left(1 + \xi\frac{y}{\tilde{\sigma}}\right)^{-1/\xi}$$

```python
import pandas as pd
import matplotlib.pyplot as plt
from pyextremes import plot_mean_residual_life
import os


data_path = f'../data/asset_analysis/{asset_name}_analysis.csv'
df = pd.read_csv(data_path, parse_dates=['date'], index_col='date')
losses = df['Loss'].dropna()

# 2. Generate the Mean-Excess (Mean Residual Life) Plot
# We will look at threshold candidates up to the 99th percentile of losses
fig, ax = plt.subplots(figsize=(10, 6))

# pyextremes handles the computation and plotting
plot_mean_residual_life(losses)

# Save the figure for your LaTeX report
os.makedirs('../report/figures', exist_ok=True)
plt.savefig(f'../report/figures/{asset_name}_mean_excess_plot.png', dpi=300, bbox_inches='tight')
plt.show()

# Print some helpful percentiles to guide our choice
print("--- Loss Percentile Reference ---")
print(f"90th Percentile: {losses.quantile(0.90):.4f}")
print(f"95th Percentile: {losses.quantile(0.95):.4f}")
print(f"98th Percentile: {losses.quantile(0.98):.4f}")
print(f"99th Percentile: {losses.quantile(0.99):.4f}")
```
